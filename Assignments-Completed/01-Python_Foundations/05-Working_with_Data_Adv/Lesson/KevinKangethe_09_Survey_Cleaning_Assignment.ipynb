```python
df["gpa"] = df["gpa"].clip(0, 4.0)
```

```python
df["gender"] = df["gender"].str.lower().map({
    "m": "Male",
    "male": "Male",
    "f": "Female",
    "female": "Female"
}).fillna("Other")
```

```python
df["dorm_resident"] = df["dorm_resident"].str.lower().map({
    "yes": True,
    "no": False
})

df["clubs_member"] = df["clubs_member"].str.lower().map({
    "yes": True,
    "no": False
})
```

```python
df["overall_academic_score"] = (
    df["gpa"] * 25 + df["attendance_rate"] * 0.75
)
```

```python
def grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    else:
        return "D"

df["grade"] = df["overall_academic_score"].apply(grade)
```
