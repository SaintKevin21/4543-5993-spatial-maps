# 🧼 Cleaning Dirty Data (Missing Values & Type Fixes)

## 🔹 LEARNING GOALS:
- Detect and count missing values (`NaN`)
- Fill or drop missing data
- Convert column data types safely
- Understand the difference between `NaN`, `None`, `""`, and type mismatches


### 🧪 1. Load a Messy Dataset

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", None],
    "Age": ["25", "thirty", 35, np.nan, "40"],
    "Signup Date": ["2022-01-01", "not a date", "2022/03/01", None, "April 5, 2022"],
    "Score": [95.5, None, 88.0, 92.5, ""]
}

df = pd.DataFrame(data)
df
```

### 🧯 2. Detecting Missing or Broken Values

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

### 🧹 3. Cleaning Strategy Options

```python
df.fillna({
    "Name": "Unknown",
    "Age": -1,
    "Signup Date": "1970-01-01",
    "Score": 0.0
})
```

```python
df.dropna()
```

### 🧬 4. Data Type Fixes

```python
df.dtypes
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Score"] = pd.to_numeric(df["Score"], errors="coerce")
df["Signup Date"] = pd.to_datetime(df["Signup Date"], errors="coerce")
df.dtypes
```

### 🩹 5. Impute (Fill In) Fixed Missing Values

```python
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Score"].fillna(df["Score"].mean(), inplace=True)
```

```python
df["Signup Date"].fillna(df["Signup Date"].min(), inplace=True)
```

```python
df["Name"].fillna("Unknown", inplace=True)
```

### 🤓 6. Cleaned Data Review

```python
df.info()
```

```python
df.describe(include="all")
```

```python
#1
print("\nMissing values (boolean map):")
print(df.isnull())

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nRows with ANY missing values:")
print(df[df.isnull().any(axis=1)])
```

```python
#2
# Convert Age and Score to numeric
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Score"] = pd.to_numeric(df["Score"], errors="coerce")

# Convert dates
df["Signup Date"] = pd.to_datetime(df["Signup Date"], errors="coerce")

print("\nData types after conversion:")
print(df.dtypes)
```

```python
#3
# Fill numeric columns
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Score"].fillna(df["Score"].mean(), inplace=True)

# Fill dates with earliest valid date
df["Signup Date"].fillna(df["Signup Date"].min(), inplace=True)

# Fill text columns
df["Name"].fillna("Unknown", inplace=True)
df["Feedback"].replace("", np.nan, inplace=True)  # treat "" as missing
df["Feedback"].fillna("No Feedback", inplace=True)
```

```python
df = pd.read_csv("data/survey.csv")
```

```python
# Drop rows that are completely empty
df.dropna(how="all", inplace=True)

# Fill numeric columns
df.fillna({
    "Age": df["Age"].median()
}, inplace=True)

# Fill text columns
df.fillna({
    "Name": "Unknown"
}, inplace=True)
```
