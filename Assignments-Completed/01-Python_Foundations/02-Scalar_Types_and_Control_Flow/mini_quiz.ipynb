### 2️⃣ Given the following snippet:

```python
x = 7
y = 15
# z = ???
```
Write a boolean statement that would print true if the value in z was between x and y.

````python
# Answer 2
x = 7
y = 15
z = 10

if z > x and z < y:
    print(True)
else:
    print(False)
````

### 3️⃣ Add one keyword to your boolean statement from above that would print false instead of true.

````python
# Answer 3
x = 7
y = 15
z = 10

if not (z > x and z < y):
    print(True)
else:
    print(False)
````

## 4️⃣ Fill in the blank to check for inequality:
```python
if score ___ 100:
    print("Perfect score!")
```

````python
# Replace ___ with the correct operator
score = 100

if score == 100:
    print("Perfect score!")
````

## 5️⃣ Given the following:

```Python
a = 11

if a % 2 == 0:
    print("Variable is even")
```

Add to the if statement so that it will also print "Variable is odd" depending on the number.

````python
# Answer 5
a = 11

if a % 2 == 0:
    print("Variable is even")
else:
    print("Variable is odd")
````
