# 📘 02 - While Loops

Learn how to repeat logic using conditions.

## ✅ Learning Goals
- Use `while` with a condition
- Understand infinite loops and loop control

```python
# 🔁 Count up to 5
counter = 1
while counter <= 5:
    print('Count:', counter)
    counter += 1
```

```python
# 🚫 Avoid infinite loop
x = 0
while x < 3:
    print('x is', x)
    x += 1
```

```python
# ⚠️ Break and continue
n = 0
while True:
    n += 1
    if n == 3:
        continue
    if n == 6:
        break
    print(n)
```
