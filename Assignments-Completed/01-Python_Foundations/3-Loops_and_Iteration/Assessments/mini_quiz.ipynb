## 2️⃣ Skipping an Iteration

Write a `for` loop that iterates over the numbers `0` through `6`.

- When the value is `3`, **skip that iteration**
- Print every other value

Your output should NOT include the number `3`.

````python
# Write your solution here
for i in range(7):
    if i == 3:
        continue
    print(i)
````

### 3️⃣ Stopping a Loop Early

Write a loop that iterates over the numbers `1` through `10`.

- Stop the loop entirely when the number `6` is reached
- Print each number **before** the loop stops

````python
# Write your solution here
for i in range(1, 11):
    if i == 6:
        break
    print(i)
````

### 5️⃣ File Iteration with `with`

Write Python code that:

- Opens a file named `data.txt` for **reading** using `with`
- Iterates over each line in the file
- Prints each line

⚠️ Assume the file exists. Do NOT hardcode file contents.

````python
# Write your solution here
with open('data.txt', 'w') as f:
    f.write("first line\n")
    f.write("second line\n")
    f.write("third line\n")
    f.write("Fourth line \n")

# Now read the file and print each line
with open('data.txt', 'r') as f:
    for line in f:
        print(line.strip())
````
